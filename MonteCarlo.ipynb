```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def monteCarlo(total, r):
    hit = 0
    x_hit = []
    y_hit = []
    x_miss = []
    y_miss = []
    n = 0
    
    for i in range(total):
        # inisiasi x dan y secara random
        rand_x = random.uniform(0,r)
        rand_y = random.uniform(0,r)
        
        # hitung x dan y
        tmp = rand_x**2 + rand_y**2
        
        # cek kondisi hit
        if tmp <= r**2:
            x_hit.append(rand_x) 
            y_hit.append(rand_y) 
            hit += 1 
        else:
            x_miss.append(rand_x) 
            y_miss.append(rand_y) 
        n += 1
        
    # hitung nilai pi
    pi = 4 * (hit / total)
        
    return pi, x_hit, y_hit, x_miss, y_miss
        
```

```python
pi, xh_arr, yh_arr, xm_arr, ym_arr = monteCarlo(5000, 1)
print("Pi: ", pi)
```

```python
plt.title("Visualisasi Monte Carlo dengan N = 5000")
plt.scatter(xh_arr, yh_arr, c='b', s=2.5)
plt.scatter(xm_arr, ym_arr, c='r', s=2.5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
arr_pi= []
sample = np.arange(1000, 20000, 500)
for i in range(len(sample)):
    for j in range(20):
        pi, xh_arr, yh_arr, xm_arr, ym_arr = monteCarlo(sample[i], 1)  
        arr_pi.append(pi)
avg_pi = np.mean(arr_pi)
```

```python
print("Rata-rata pi: ",avg_pi)
```

```python
plt.figure(figsize=(8,6))
plt.title("Nilai Estimasi Pi terhadap Ukuran Sampel")
plt.plot(arr_pi, '.', c='b')
plt.axhline(avg_pi, c='r', label="average pi", linewidth=3)
plt.axhline(3.14, c='g', label="exact pi", linewidth=3)
plt.legend()
plt.show()
```
